```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def PegarNoticiaDaAPI(dia,exportacao): 
    # Fazendo a requisição para a API do NewsAPI
    api_key = 'fac1628b04464126940239f493d9eda5'
    noticia = requests.get(f'https://newsapi.org/v2/everything?q=tesla&from=2024-09-{dia}&sortBy=publishedAt&language=en&apiKey={api_key}')

    if noticia.status_code != 200:
        print(f"Erro na requisição: {noticia.status_code}")
        print(f'Não foi possível pegar noticias para o dia {dia}')
        return
    else:
        # Convertendo a resposta em JSON
        noticia_json = noticia.json()

        # Inicializando a lista para armazenar os dados das notícias
        dados_noticia = []

        # Função para fazer scraping do conteúdo completo
        def get_full_content(url):
            response = requests.get(url)
            soup = BeautifulSoup(response.content, 'html.parser')  # Usando o parser html.parser
            paragraphs = soup.find_all('p')
            full_text = ' '.join([para.get_text() for para in paragraphs])
            return full_text

        # Inicializando o contador
        i = 0

        # Iterando para coletar dados das primeiras 10 notícias
        while i < 10 and i < len(noticia_json['articles']):  # Certificando-se de não ultrapassar o número de artigos disponíveis
            article = noticia_json['articles'][i]
            full_content = get_full_content(article['url'])
            word_tesla_count = 'Tesla'
            word_elon_musk_count = 'Elon Musk'
            word_car_count = 'Car'  # Palavra que você quer contar
            word_contada_tesla = full_content.lower().count(word_tesla_count.lower())
            word_contada_elon_mushem = full_content.lower().count(word_elon_musk_count.lower())
            word_contada_car = full_content.lower().count(word_car_count.lower())  # Contando a palavra (case insensitive)
            dici_noticia = {
                'Autor': article['author'],
                'Titulo': article['title'],
                'Descricao': article['description'],
                'Conteudo': full_content,
                'Contagem Palavra Tesla': word_contada_tesla,
                'Contagem Palavra Elon Mushen': word_contada_elon_mushem,
                'Contagem Palavra Carro': word_contada_car  # Adicionando a contagem da palavra no dicionário
            }
            dados_noticia.append(dici_noticia)
            i += 1

        # Criando um DataFrame com os dados das notícias
        df = pd.DataFrame(dados_noticia)

        # Exibindo o DataFrame
        display(df)

        if exportacao == 'csv':
            # Exportando o DataFrame para um arquivo CSV
            df.to_csv(f'noticias_tesla_dia_{dia}.csv', index=False)
        elif exportacao == 'json':
            # Exportando o DataFrame para um arquivo JSON
            df.to_json(f'noticias_tesla_dia_{dia}.json', orient='records', lines=True)

```

```python
PegarNoticiaDaAPI(12,'json')
```
